```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('Mall_customers.csv')
X = dataset.iloc[:, [3, 4]].values
```

```python
dataset.head()
```

```python
X
```

```python
from sklearn.cluster import KMeans

wcss = []

for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, 
                    init = 'k-means++', 
                    n_init = 10, 
                    max_iter = 300,
                    random_state = 0)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
    
plt.plot(range(1, 11), wcss)
plt.title('The Elbow Method Graph')
plt.xlabel('wcss')
plt.ylabel('Number of clusters')
plt.show()
```

```python
## applying the KMean model to the dataset

kmeans = KMeans(n_clusters = 5,
               init = 'k-means++',
               n_init = 10,
               max_iter = 300,
               random_state = 0)
y_kmeans = kmeans.fit_predict(X)
y_kmeans
```

```python
plt.scatter(X[y_kmeans == 0, 0], X[y_kmeans == 0, 1], s = 50, c = 'red', label = ' Carefull')
plt.scatter(X[y_kmeans == 1, 0], X[y_kmeans == 1, 1], s = 50, c = 'blue', label = 'Standerd')
plt.scatter(X[y_kmeans == 2, 0], X[y_kmeans == 2, 1], s = 50, c = 'green', label = 'Target')
plt.scatter(X[y_kmeans == 3, 0], X[y_kmeans == 3, 1], s = 50, c = 'cyan', label = 'Careless')
plt.scatter(X[y_kmeans == 4, 0], X[y_kmeans == 4, 1], s = 50, c = 'magenta', label = 'Sensible')

plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s = 150, c = 'yellow', label = 'Centroids')

plt.title('Cluster Of Customers')
plt.xlabel('Annual Icome (K$)')
plt.ylabel('Spending Score (1-100)')
plt.legend()
plt.show()

```
